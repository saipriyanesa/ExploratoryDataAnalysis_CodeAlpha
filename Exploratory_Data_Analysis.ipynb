```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Month':['Jan','Feb','Mar','Apr','May','Jun',
             'Jul','Aug','Sep','Oct','Nov','Dec'],
    'Sales':[12000,15000,18000,22000,25000,28000,
             30000,27000,26000,31000,34000,38000],
    'Profit':[3000,4000,5000,6500,7000,8000,
              9000,8500,7800,9500,11000,12500]
}

df = pd.DataFrame(data)

df
```

```python
print("Dataset Information")
df.info()
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
print("Highest Sales:", df['Sales'].max())
print("Lowest Sales:", df['Sales'].min())

print("Highest Profit:", df['Profit'].max())
print("Lowest Profit:", df['Profit'].min())
```

```python
best_month = df.loc[df['Sales'].idxmax()]

print("Best Performing Month")
print(best_month)
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Sales'],
         marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Profit'],
         marker='o')

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Sales'],
             bins=6,
             kde=True)

plt.title("Sales Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Sales',
    y='Profit',
    data=df,
    s=100
)

plt.title("Sales vs Profit")

plt.show()
```

```python
correlation = df[['Sales','Profit']].corr()

print(correlation)
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Sales'] < lower) |
              (df['Sales'] > upper)]

print("Outliers:")
print(outliers)
```

```python
print("EDA Findings")

print("- Sales generally increase over time.")
print("- December has the highest sales and profit.")
print("- Strong positive relationship exists between sales and profit.")
print("- No missing values found.")
print("- No significant outliers detected.")
```
